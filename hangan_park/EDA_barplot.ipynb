## **1. Barplot**

### **1.1 접근성 정보**
  - 버스, 지하철
    - 'bus_stop', 'bus_line', 'bus_walk_min', 'sub_station','sub_line', 'sub_walk_min', 'sub_walk_mean'

```python
df = df.sort_values('hangang', ascending=False).reset_index(drop=True)

columns = ['bus_stop', 'bus_line', 'bus_walk_min', 'sub_station','sub_line', 'sub_walk_min', 'sub_walk_mean']
num_columns = len(columns)

titles = ['버스 정류장 개수', '버스 노선 수', '정류장까지의 최소 도보 거리', '지하철 역 개수', '지하철 노선 개수', '역까지의 최소 도보 거리', '역까지의 평균 도보 거리']

# 필요한 subplot 수 계산
num_rows = (num_columns + 3) // 4

# subplot을 생성
fig, axs = plt.subplots(num_rows, 4, figsize=(20, 5 * num_rows))

# axs를 1차원 배열로 평면화
axs = axs.flatten()

# 공원별 고유 색상 지정
color_map = {
    '강서': '#980000',
    '광나루': '#CC3D3D',
    '난지': '#CC723D',
    '뚝섬': '#CCA63D',
    '망원': '#9FC93C',
    '반포': '#47C83E',
    '양화': '#3DB7CC',
    '여의도': '#4174D9',
    '이촌': '#4641D9',
    '잠실': '#7E41D9',
    '잠원': '#D941C5'
}

for i, (column, title) in enumerate(zip(columns, titles)):
    for idx, area in enumerate(df['hangang']):
        color = color_map.get(area, '#808080')
        axs[i].barh(area, df[column][idx], color=color, alpha=0.7)
    axs[i].set_title(f'{title}', fontsize=20)
    axs[i].tick_params(axis='y', labelsize=12)

# 남은 빈 subplot 제거
for j in range(i + 1, num_rows * 4):
    fig.delaxes(axs[j])

plt.tight_layout()
plt.show()
```
