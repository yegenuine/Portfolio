- Convert `total_cost` to `avg_cost`

```python
# Function to calculate the number of days in a quarter
def days_in_quarter(year_quarter):
    year = int(str(year_quarter)[:4])
    quarter = int(str(year_quarter)[4])
    if quarter == 1:
        return pd.Timestamp(f"{year}-04-01") - pd.Timestamp(f"{year}-01-01")
    elif quarter == 2:
        return pd.Timestamp(f"{year}-07-01") - pd.Timestamp(f"{year}-04-01")
    elif quarter == 3:
        return pd.Timestamp(f"{year}-10-01") - pd.Timestamp(f"{year}-07-01")
    elif quarter == 4:
        return pd.Timestamp(f"{year+1}-01-01") - pd.Timestamp(f"{year}-10-01")
```

```python
# Apply the function to calculate days in quarter
income_data['days_in_quarter'] = income_data['quarter'].apply(days_in_quarter).dt.days

# Divide the costs by the number of days in the quarter
income_data['drink_cost'] = income_data['drink_cost'] / income_data['days_in_quarter']
income_data['leisure_cost'] = income_data['leisure_cost'] / income_data['days_in_quarter']
income_data['food_cost'] = income_data['food_cost'] / income_data['days_in_quarter']

# Drop the auxiliary column
income_data.drop(columns=['days_in_quarter'], inplace=True)
income_data
```

- Convert `quarter` to `Y-m-d`
    - `merged_data`에 `quarter` 변수를 생성한 후 `adm_cd`, `quarter`를 기준으로 `income_data` 병합

```python
# Function to convert date to corresponding quarter value
def date_to_quarter(date):
    if date.month in [1, 2, 3]:
        return f"{date.year}1"
    elif date.month in [4, 5, 6]:
        return f"{date.year}2"
    elif date.month in [7, 8, 9]:
        return f"{date.year}3"
    elif date.month in [10, 11, 12]:
        return f"{date.year}4"

# Convert DATE column to datetime format
merged_data['DATE'] = pd.to_datetime(merged_data['DATE'])

# Apply the function to create a new quarter column
merged_data['quarter'] = merged_data['DATE'].apply(date_to_quarter)
merged_data
```

- Convert `object` to `int`

```python
merged_data['quarter'] = merged_data['quarter'].astype('int64')
merged_data.info()
```
